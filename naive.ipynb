Naive Bayes Algorithm

Why?-To check the probability of particular event

Independent Events: Rolling a dice{1 ,2,3,4,5,6}
p(1)=1/6 p(2)=1/6 p(3)=1/6 and so on

Dependent Event: 
 
If there are 3 orange and 2 yellow coins

then, p(o)=3/5
now if we remove  a orange coin then p(y/o)=3/5
p(o and y)=p(o)*p(o/y)

p(A and B)=p(A)*p(B/A)

p(A and B)=p(B and A)
p(A)*P(B/A)=p(B)*p(A/B)
p(A/B)=p(A)*P(B/A)/(p(B))
The above statement denotes Bayes Theorem

p(A/B)=Probability of event A given B event occurred
similarly for P(B/A)

p(y/x1x2x3)=p(y)*p(x1,x2,x3/y)/(p(x1,x2,x3))
           =(p(y)*p(x1/y)*p(x2/y)*p(x3/y))/p(x1,x2,x3)

p(yes/x1,x2,x3)=(p(yes)*p(x1/yes)*p(x2/yes)*p(x3/yes))/p(x1,x2,x3)
               =1/2*p(x1/yes)*p(x2/yes)*p(x3/yes))/p(x1,x2,x3)

p(No/x1,x2,x3)=(1/2*p(x1,x2,x3/No))/p(x1,x2,x3)

p(yes/x1,x2,x3)=p(No/x1,x2,x3)
p(x1/yes)*p(x2/yes)*p(x3/yes))=p(x1,x2,x3/No)

EXAMPLE-1:

test(sunny,hot)-> Play Tennis(yes/No)
pt(yes)=9/14
pt(no)=5/14

p(hot,yes)=2/9
p(hot,no)=2/5


p(sunny,yes)=2/9
p(sunny,no)=3/5



p(yes/sunny,hot)=9/14*2/9*2/9=0.0317
p(no/sunny,hot)=5/14*3/5*2/5=0.0857

p(yes)=p(yes)/P(yes)+p(no)=0.27==27%
p(no)=p(no)/p(no)+p(yes)=0.73=73%

so we can say that 73 % people cannot play Tennis and 27% can play.
